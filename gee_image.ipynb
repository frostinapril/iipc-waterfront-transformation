```python
coords = [
    [xmin, ymin],
    [xmax, ymin],
    [xmax, ymax],
    [xmin, ymax],
    [xmin, ymin]  # close polygon
]
region = ee.Geometry.Polygon([coords])
```

```python
s2 = (ee.ImageCollection('COPERNICUS/S2_SR')
      .filterBounds(region)
      .filterDate('2023-06-01', '2023-08-31')  # summer 2025
      .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 10))  # <10% clouds
      .sort('CLOUDY_PIXEL_PERCENTAGE')
     )

# Select first image (least clouds)
img = s2.first()

# Select bands: Red, Green, Blue, NIR
# Sentinel-2 SR band names: B4=Red, B3=Green, B2=Blue, B8=NIR, B11=SWIR
img_rgbnir = img.select(['B4', 'B3', 'B2', 'B8', 'B11'])

# 4. Export to Google Drive
task = ee.batch.Export.image.toDrive(
    image=img_rgbnir,
    description='roosendaal2023',
    folder='GEE_Exports',      # folder in your Google Drive
    fileNamePrefix='roosendaal2023',
    region=region,
    scale=10,                  # Sentinel-2 10m resolution
    crs='EPSG:28992',          # Amersfoort / RD New
    maxPixels=1e13              # very large images
)
task.start()

print("Export task started. Check Google Drive folder 'GEE_Exports'.")
```
