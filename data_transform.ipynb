```python
import json
import requests

# Get the data
url = "https://connections.routescanner.com/rotterdam.data?connectionType=export&modalities=BARGE"
response = requests.get(url)
data = response.json()

print(data)
```

```python
index1 = data.index("coordinates")
ref_route = data[index1 +1]
ref_list = []
for ref in ref_route:
    ref_list.append(data[ref])

ref_list_2 = []
for ref in ref_list:
    lat = data[ref[0]]
    lon = data[ref[1]]
    ref_list_2.append([lat, lon])

coordinate_list = []
for ref in ref_list_2:
    lat_1 = data[ref[0][0]]
    lon_1 = data[ref[0][1]]
    lat_2 = data[ref[1][0]]
    lon_2 = data[ref[1][1]]
    coordinate_list.append("LINESTRING({} {}, {} {})".format(lat_1, lon_1, lat_2, lon_2))

with open("routes.wkt", "w") as f:
    for coordinate in coordinate_list:
        f.write(coordinate)
        f.write("\n")
```
